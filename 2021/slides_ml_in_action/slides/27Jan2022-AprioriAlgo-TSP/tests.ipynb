## To see if the helper functions are working

```python
import apriori
```

```python
data = apriori.load_data()
print("THe synthetic dataset of transactions we will be using to do a test is: ",data)
```

```python
c1 = apriori.create_ci1(data)
print("Candidate itemsets of size 1 from the data are: ", c1)
```

```python
D = list(map(set, data))
print("The list of transactions in the dataset are: ",D)
```

```python
l1, support_data0 = apriori.scan_data(D, c1, 0.5)
print("Frequent itemsets of size 1 above threshold of 50% are: ",l1)
print()
print("Their corresponding support values are given as follows: ",support_data0)
```
